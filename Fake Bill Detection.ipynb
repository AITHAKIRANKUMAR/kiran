```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('Fake_Bill.csv')
df_numeric = df.drop(columns=["is_genuine"])  # Exclude categorical column

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_numeric)
plt.xticks(rotation=45)
plt.title("Boxplot of Numerical Features")
plt.show()

df_numeric.hist(figsize=(12, 8), bins=30, edgecolor='black')
plt.suptitle("Histogram of Numerical Features")
plt.show()

outliers = {}
for column in df_numeric.columns:
    Q1 = df_numeric[column].quantile(0.25)
    Q3 = df_numeric[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers[column] = df_numeric[(df_numeric[column] < lower_bound) | (df_numeric[column] > upper_bound)][column].count()

print("Outlier Counts Per Feature:")
for key, value in outliers.items():
    print(f"{key}: {value} outliers")
```
